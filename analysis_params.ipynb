### Normalize 

```python
def normalize_dataframe(dataf):
    maxval = max(dataf.max())
    minval = min(dataf.min())
    dataf0 = dataf.copy()
    for col in dataf.columns:
        dataf0[col] = (dataf0[col]-minval)/(maxval-minval)


    return dataf0   
```
